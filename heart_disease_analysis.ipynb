```python
# GRAFICOS DE COLESTEROL 
```

```python
sns.violinplot(x='restecg',y='chol',data=hd,hue='sex',palette='viridis') # COLESTEROL POR TIPO DE ECG Y GENERO
```

```python
sns.scatterplot(x='trestbps',y='chol',data=hd,hue='sex') # SCATTER DE TRESTBPS POR COLESTEROL 
```

```python

plt.figure(figsize=(10,6))

sns.countplot(data=hd, x='thal', hue='cp') # GRAFICO HALLAZGO ( PACIENTES ASINTOMATICOS CON ISTEMIA)

plt.title('CP OVER THAL DISTRIBUTION')
plt.xticks(rotation=30)
plt.show()
```

```python
# HEATMAP DE CORRELACION ENTRE VARIABLES NUMERICAS
```

```python
corr = hd.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,6)) 
sns.heatmap(corr,annot=True,cmap='viridis')
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='slope',y='oldpeak',data=hd,hue='exang',palette='viridis') # VALORES DE OLDPEAK SEGUN SLOPE Y EXANG
```
